```python
# 3. OpenCV로 영상 재생
import cv2
from IPython.display import display, Image

# 영상 파일 경로 설정
video_path = os.path.join(project_path, 'dance.mov')

# OpenCV를 사용하여 영상 재생
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")

else:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # cv2.imshow(frame)
        # 로컬에서 코드를 실행시킬 때는 위 코드를 사용해 간단히 프레임을 재생할 수 있다

        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    cap.release()
```
